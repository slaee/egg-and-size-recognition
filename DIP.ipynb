```python
from skimage import io
import numpy as np
from PIL import Image

photo = io.imread('./datasets/tests/1-52.4-48722.png')
type(photo)
```

```python
# show the gray image
plt.imshow(photo, cmap='gray')
```

```python
photo_2d = photo[:,:]
threshold = np.mean(photo_2d)

above_threshold = photo_2d > threshold
photo_2d[above_threshold] = 255
photo_2d[~above_threshold] = 0

count = np.sum(above_threshold)

plt.imshow(photo_2d, cmap='gray')
print (f"size: {count}")
```

```python
# convert grayscale to np floating point
# normalize the image
gray_scale_2d_norm = (photo_2d - photo_2d.min()) / (photo_2d.max() - photo_2d.min())
```
