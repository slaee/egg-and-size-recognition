```python
from skimage import io
import numpy as np
from PIL import Image

photo = io.imread('./datasets/egg.png')
type(photo)
```

```python
import matplotlib.pyplot as plt
plt.imshow(photo)
```

```python
gray_scale_photo = np.mean(photo, axis=2)

# convert the 2D numpy array to 3D numpy array
# so using np.stack to create 3D numpy array by stacking gray_scale_photo along with third axis -1 three times
# np.uint8 to ensure the photo will be represented in 8bits per channel to allow range of 256 for grayscale
gray_scale_photo = np.stack([gray_scale_photo]*3, axis=-1).astype(np.uint8)

# using the method above for mirror image
plt.imshow(gray_scale_photo[::1])
```

```python
gray_scale_photo_2d = np.mean(photo, axis=2).astype(np.uint8)
threshold = np.mean(gray_scale_photo_2d)

above_threshold = gray_scale_photo_2d > threshold
gray_scale_photo_2d[above_threshold] = 255
gray_scale_photo_2d[~above_threshold] = 0

count = np.sum(above_threshold)

plt.imshow(gray_scale_photo_2d, cmap='gray')
print (f"size: {count}")
```
